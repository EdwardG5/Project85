```python
import matplotlib.pyplot as plt
import numpy as np

# Create the vectors X and Y
x = np.array(range(10000))
y = np.array([int(_x).bit_length() + 3 + 1 for _x in x]) # +1 at end as otherwise net benefit = 0

# Create the plot
plt.plot(x,y,label='y = bit_length(x) + 3 + 1')

# Add a title
plt.title('Number of k-mers in reference vs. min size of k-mer')

# Add X and y Label
plt.xlabel('Number of k-mers in Reference')
plt.ylabel('Minimum size of k-mer')

# Add a grid
plt.grid(alpha=.4,linestyle='--')

# Add a Legend
plt.legend()

# Show the plot
plt.show()
```
